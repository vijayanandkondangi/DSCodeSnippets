```python
loans = pd.read_csv("loans.csv")
```

```python
# drop rows containing NA

loans = loans.dropna()
```

```python
loans.count()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test = train_test_split(loans, test_size=0.3, random_state = 123)
```

```python
X_train1 = X_train[(X_train["notfullypaid"] == 1)]
X_train0 = X_train[(X_train["notfullypaid"] == 0)]
```

```python
from sklearn.model_selection import train_test_split
X_train080, X_train020 = train_test_split(X_train0, test_size=0.2, random_state = 123)
```

```python
X_train1 = X_train1.append(X_train020)
```

```python
X_train = X_train1
```

```python
X_train["notfullypaid"].value_counts()
```

```python
# compose a list of dependent variable

y_train = X_train["notfullypaid"]
y_test = X_test["notfullypaid"]
```

```python
# drop the dependent variable column

X_train = X_train.drop(["notfullypaid"], axis = 1)
X_test = X_test.drop(["notfullypaid"], axis = 1)
```

```python
# baseline accuracy of the dataset (84.02%)

loans["notfullypaid"].value_counts()
```

```python
# converting factor in to dummy variable

x_trainpurpose = pd.get_dummies(X_train["purpose"],drop_first=True)
x_testpurpose = pd.get_dummies(X_test["purpose"],drop_first=True)

X_train = X_train.drop(["purpose"], axis = 1)
X_test = X_test.drop(["purpose"], axis = 1)

X_train = pd.concat([X_train,x_trainpurpose],axis=1)
X_test = pd.concat([X_test,x_testpurpose],axis=1)
```

```python
X_train.head(5)
```

```python
from sklearn.metrics import cohen_kappa_score, make_scorer
kappa_scorer = make_scorer(cohen_kappa_score)
```

```python
# Logistic Regression Model using GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
Cs = [0.0001, 0.001, 0.01, 0.1, 1, 10]
param_grid = {'C': Cs, 'penalty':["l1"]}
grid_search = GridSearchCV(LogisticRegression(), param_grid, verbose=3, cv=5, scoring=kappa_scorer)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.score(X_test, y_test)
```

```python
# Only a marginal increase in baseline accuracy was obtained
```

```python
# Decision tree model with GridSearchCV

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
depths = [1, 2, 3, 4, 5]
param_grid = {'max_depth': depths}
grid_searchDT = GridSearchCV(DecisionTreeClassifier(), param_grid, verbose=3, cv=5, scoring=kappa_scorer)
grid_searchDT.fit(X_train, y_train)
```

```python
grid_searchDT.best_params_
```

```python
grid_searchDT.best_score_
```

```python
grid_searchDT.score(X_test, y_test)
```

```python
# Only a marginal increase in baseline accuracy was obtained
```

```python
# Random Forest model using GridSearchCV

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
estimators = [50, 100, 150, 200, 250]
param_grid = {'n_estimators': estimators}
grid_searchRF = GridSearchCV(RandomForestClassifier(), param_grid, verbose=3, cv=5, scoring=kappa_scorer)
grid_searchRF.fit(X_train, y_train)
```

```python
grid_searchRF.best_params_
```

```python
grid_searchRF.best_score_
```

```python
grid_searchRF.score(X_test, y_test)
```

```python
# GB Classifier model using GridSearchCV

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
estimators = [25, 50, 75, 100, 125]
depth = [1, 2, 3, 4]
rate = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
param_grid = {'n_estimators': estimators, 'max_depth' : depth, 'learning_rate' : rate}
grid_searchGBC = GridSearchCV(GradientBoostingClassifier(), param_grid, verbose=3, cv=5, scoring=kappa_scorer)
grid_searchGBC.fit(X_train, y_train)
```

```python
grid_searchGBC.best_score_
```

```python
grid_searchGBC.score(X_test, y_test)
```

```python
# SVC

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
Gs = [0.001, 0.01, 0.1, 1, 10]
Cs = [0.001, 0.01, 0.1, 1, 10]
param_grid = {'C': Cs, 'gamma' : Gs}
grid_searchSVC = GridSearchCV(SVC(probability = True), param_grid, verbose=3, cv=5, scoring=kappa_scorer)
grid_searchSVC.fit(X_train, y_train)
```

```python
def truth(item):
    if item > 0.4625:
        return 0
    else:
        return 1

predprob = grid_searchSVC.predict_proba(X_test)
pred = [truth(float(x)) for x in predprob[:,[0]]]
from sklearn.metrics import confusion_matrix 
confusion = confusion_matrix(y_test, pred)
print(confusion)

# threshold = 0.5, error rate = 16.39%
# threshold = 0.4, error rate = 15.72%
# threshold = 0.3, error rate = 15.94%
# threshold = 0.25, error rate = 15.62%
```
